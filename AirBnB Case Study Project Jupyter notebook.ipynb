```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('files/AB_NYC_2019.csv')
air_bnb.head(20)
```

```python
# Which hosts are the busiest and why?
# print(air_bnb[['host_id', 'availability_365']])
Avail0 = air_bnb.query('availability_365 == 0')
Min = Avail0.query('minimum_nights <= 2')
Min.query('calculated_host_listings_count > 4')

These 137 hosts have high property counts and the busiest turn around. 
```

```python
# How many neighborhood groups are available and which shows up the most?
pd.unique(air_bnb['neighbourhood_group'])


Manhattan shows up the most of the 5 available neighbourhoods.
```

```python
air_bnb['neighbourhood_group'].value_counts()
```

```python
# Are private rooms the most popular in manhattan?
Neighbourhood = air_bnb.query('neighbourhood_group == "Manhattan"')
Ans = Neighbourhood['room_type'].value_counts()
Ans

There are more Entire home/apts in Manhattan
```

```python
# Which hosts are the busiest based on their reviews?
Num = air_bnb.query('number_of_reviews > 100')
Month = Num.query('reviews_per_month >= 1')
Ans = Month.query('availability_365 >=0') 
Ans
The hosts with the most reviews and the most per month who are fully booked. 
```

```python
#Which neighorhood group has the highest average price?
avg = air_bnb.groupby('neighbourhood_group', as_index=False)['price'].mean()
avg
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby('neighbourhood_group', as_index=False)['price'].max()

Several of them
```

```python
#Which top 5 hosts have the highest total price?
top_5 = air_bnb.groupby('host_id')['price'].max()
top_5.sort_values(ascending=False)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
no_avail = air_bnb.query('availability_365 == 0')
count = no_avail.query('number_of_reviews >= 100')
count
```

```python
# What host has the highest total of prices and where are they located?
host = air_bnb.query('price >= 7500')
host

There are 3 hosts with the highest listing prices, one in Queens, one in Manhattan, and one in Brooklyn
```

```python
# When did Danielle from Queens last receive a review?
name = air_bnb.loc[air_bnb['host_name'] == 'Danielle']
group = name.loc[name['neighbourhood_group'] == 'Queens']
group['last_review'].sort_values(ascending=False)

There are several Hosts named Danielle in Queens but the last one to recieve a review was host_id=22469
who last recieved a review on 2019-07-08
```
